```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import re
import math
```

```python
file_path = r'E:\暨南大学\碳排放与碳中和\Proposal\Biofouling\Paper\Test Results\Trajectory\\'

file_name_1 = ['Random_Sim_4_Info_normal_30_10.npy','Random_Sim_3_Info_normal_100_50.npy','Random_Sim_7_Info_normal_200_80.npy',
               'Random_Sim_2_Info_normal_300_100.npy','Random_Sim_6_Info_normal_400_200.npy','Random_Sim_1_Info_normal_500_150.npy']
file_name_2 = ['Random_Sim_4_Depth_normal_30_10.npy','Random_Sim_3_Depth_normal_100_50.npy','Random_Sim_7_Depth_normal_200_80.npy',
               'Random_Sim_2_Depth_normal_300_100.npy','Random_Sim_6_Depth_normal_400_200.npy','Random_Sim_1_Depth_normal_500_150.npy']

dep_10 = []
dep_50 = []
dep_100 = []
dep_200 = []
dep_500 = []
dep_over = []

depth_data = np.load(file_path+file_name_2[5]) 
info_data = np.load(file_path+file_name_1[5])
for i in range(len(info_data)):
    size_t = int((float(info_data[i][7])*3/(4*math.pi))**(1/3)*2*1000000)
    if size_t < 10:
        dep_10.append(depth_data[i])
    if size_t < 50 and size_t >= 10:
        dep_50.append(depth_data[i])
    if size_t < 100 and size_t >= 50:
        dep_100.append(depth_data[i])
    if size_t < 500 and size_t >= 100:
        dep_200.append(depth_data[i])
    if size_t < 1000 and size_t >= 500:
        dep_500.append(depth_data[i])
    if size_t >= 1000:
        dep_over.append(depth_data[i])
```

```python
file_name_3 = ['Random_Sim_4_Info_normal_30_10.npy','Random_Sim_2_Info_normal_300_100.npy','Random_Sim_1_Info_normal_500_150.npy']
file_name_4 = ['Random_Sim_4_Depth_normal_30_10.npy','Random_Sim_2_Depth_normal_300_100.npy','Random_Sim_1_Depth_normal_500_150.npy']
time_t = ['30 days', '300 days', '500 days']
col = ['blue','green','red']

plt.figure(figsize=(20,20),dpi=500)
grid = plt.GridSpec(3,3,wspace=0.2,hspace=0.2)

font1 = {'family': 'Arial',
            'weight': 'normal',
            'size': 20,
            }

font2 = {'family': 'Arial',
            'weight': 'normal',
            'size': 20,
            }

density = ['(A) 900-930 kg/$\mathregular{m^3}$', '(B) 930-960 kg/$\mathregular{m^3}$', '(C) 960-1000 kg/$\mathregular{m^3}$']
shape = ['(D) Sphere', '(E) Fiber', '(F) Film']
subtitle = ['(G) 0-200 um','(H) 200-500 um','(I) Over 500 um']

for i in range(len(file_name_3)):
    dep_930 = []
    dep_960 = []
    dep_1000 = []

    depth_data = np.load(file_path+file_name_4[i]) 
    info_data = np.load(file_path+file_name_3[i])
    for j in range(len(info_data)):
        if int(info_data[j][9]) < 938:
            dep_930.append(depth_data[j])
        if int(info_data[j][9]) >= 938 and int(info_data[j][10]) < 952:
            dep_960.append(depth_data[j])
        if int(info_data[j][9]) >= 952:
            dep_1000.append(depth_data[j])
    
    dep_re = [dep_930,dep_960,dep_1000]

    plt.subplot(grid[0:1,0:1])        
    if i == 0:
        plt.title(density[0],fontsize = 22)
    num_bin = int(100 * np.max(dep_re[0])/1000)
    a = plt.hist(dep_re[0],bins=num_bin,orientation='horizontal',color = 'white')
    plt.plot(a[0],a[1][1:],color = col[i],linewidth = 3.0)
    plt.ylabel('Depth (m)',font2)
    plt.tick_params(labelsize = 18)
    plt.ylim(1000, 0)
    #plt.text(max(a[0])*0.8, 950, time_t[i], ha='center', fontsize=20)

    plt.subplot(grid[0:1,1:2])
    if i == 0:
        plt.title(density[1],fontsize = 22)
    num_bin = int(100 * np.max(dep_re[1])/1000)
    a = plt.hist(dep_re[1],bins=num_bin,orientation='horizontal',color = 'white')
    plt.plot(a[0],a[1][1:],color = col[i],linewidth = 3.0)
    
    plt.tick_params(labelsize = 18)
    plt.ylim(1000, 0)

    plt.subplot(grid[0:1,2:3])
    if i == 0:
        plt.title(density[2],fontsize = 22)
    num_bin = int(100 * np.max(dep_re[2])/1000)
    a = plt.hist(dep_re[2],bins=num_bin,orientation='horizontal',color = 'white')
    plt.plot(a[0],a[1][1:],color = col[i],linewidth = 3.0)
    plt.tick_params(labelsize = 18)
    plt.ylim(1000, 0)

for i in range(len(file_name_3)):
    dep_sphere = []
    dep_fiber = []
    dep_film = []

    depth_data = np.load(file_path+file_name_4[i]) 
    info_data = np.load(file_path+file_name_3[i])
    for j in range(len(info_data)):
        if info_data[j][0] == 'sphere':
            dep_sphere.append(depth_data[j])
        if info_data[j][0] == 'fiber':
            dep_fiber.append(depth_data[j])
        if info_data[j][0] == 'film':
            dep_film.append(depth_data[j])
    
    dep_re = [dep_sphere,dep_fiber,dep_film]

    plt.subplot(grid[1:2,0:1])        
    if i == 0:
        plt.title(shape[0],fontsize = 22)
    num_bin = int(100 * np.max(dep_re[0])/1000)
    a = plt.hist(dep_re[0],bins=num_bin,orientation='horizontal',color = 'white')
    plt.plot(a[0],a[1][1:],color = col[i],linewidth = 3.0)
    plt.ylabel('Depth (m)',font2)
    plt.tick_params(labelsize = 18)
    plt.ylim(1000, 0)
    plt.legend(labels=["30th day","200th day","500th day"], loc = 'lower right', fontsize=18)
    #plt.text(max(a[0])*0.8, 950, time_t[i], ha='center', fontsize=20)

    plt.subplot(grid[1:2,1:2])
    if i == 0:
        plt.title(shape[1],fontsize = 22)
    num_bin = int(100 * np.max(dep_re[1])/1000)
    a = plt.hist(dep_re[1],bins=num_bin,orientation='horizontal',color = 'white')
    plt.plot(a[0],a[1][1:],color = col[i],linewidth = 3.0)
    plt.tick_params(labelsize = 18)
    plt.ylim(1000, 0)

    plt.subplot(grid[1:2,2:3])
    if i == 0:
        plt.title(shape[2],fontsize = 22)
    num_bin = int(100 * np.max(dep_re[2])/1000)
    a = plt.hist(dep_re[2],bins=num_bin,orientation='horizontal',color = 'white')
    plt.plot(a[0],a[1][1:],color = col[i],linewidth = 3.0)
    plt.tick_params(labelsize = 18)
    plt.ylim(1000, 0)
    
for i in range(len(file_name_3)):
    dep_200 = []
    dep_500 = []
    dep_over = []

    depth_data = np.load(file_path+file_name_4[i]) 
    info_data = np.load(file_path+file_name_3[i])
    for j in range(len(info_data)):
        size_t = int((float(info_data[j][7])*3/(4*math.pi))**(1/3)*2*1000000)
        if size_t < 200:
            dep_200.append(depth_data[j])
        if size_t < 500 and size_t >= 200:
            dep_500.append(depth_data[j])
        if size_t >= 500:
            dep_over.append(depth_data[j])
    
    dep_re = [dep_200,dep_500,dep_over]

    plt.subplot(grid[2:3,0:1])        
    if i == 0:
        plt.title(subtitle[0],fontsize = 22)
    num_bin = int(100 * np.max(dep_re[0])/1000)
    a = plt.hist(dep_re[0],bins=num_bin,orientation='horizontal',color = 'white')
    plt.plot(a[0],a[1][1:],color = col[i],linewidth = 3.0)
    plt.ylabel('Depth (m)',font2)
    if i == 2:
        plt.xlabel('particles/$\mathregular{m^3}$',font2)
    plt.tick_params(labelsize = 18)
    plt.ylim(1000, 0)
    #plt.text(max(a[0])*0.8, 950, time_t[i], ha='center', fontsize=20)

    plt.subplot(grid[2:3,1:2])
    if i == 0:
        plt.title(subtitle[1],fontsize = 22)
    num_bin = int(100 * np.max(dep_re[1])/1000)
    a = plt.hist(dep_re[1],bins=num_bin,orientation='horizontal',color = 'white')
    plt.plot(a[0],a[1][1:],color = col[i],linewidth = 3.0)
    if i == 2:
        plt.xlabel('particles/$\mathregular{m^3}$',font2)
    plt.tick_params(labelsize = 18)
    plt.ylim(1000, 0)

    plt.subplot(grid[2:3,2:3])
    if i == 0:
        plt.title(subtitle[2],fontsize = 22)
    num_bin = int(100 * np.max(dep_re[2])/1000)
    a = plt.hist(dep_re[2],bins=num_bin,orientation='horizontal',color = 'white')
    plt.plot(a[0],a[1][1:],color = col[i],linewidth = 3.0)
    if i == 2:
        plt.xlabel('particles/$\mathregular{m^3}$',font2)
    plt.tick_params(labelsize = 18)
    plt.ylim(1000, 0)

plt.show()
```
